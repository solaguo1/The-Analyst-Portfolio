```python
#Q2 Is the race slower in the summer than the winter?
```

```python
#try to split by delmiter . and get get the month

df3['Race_Month']=df3['Race_Day'].str.split('.').str.get(1).astype(int)
```

```python
# apply lambda x and the if statment to determine the season 

df3['Race_Season']= df3['Race_Month'].apply(lambda x:'Winter' if x>11 else 'Fall'if x>8 else 'Summer' if x>5 else 'spring' if x>2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('Race_Season')['Athlete_Average_Speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
# It is easiest to run the race in spring, in the summer the racer run slower for almost 1km per hour. Possible due to the heat
# Summer is also the season with the least amount of racer, possibly due to the heat less people would like to participate and less race take place. 
```
